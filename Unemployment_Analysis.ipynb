Estimated Labour Participation Rate (%) --> measure that represents the proportion of the working-age population that is either
                                    employed or actively seeking employment

                                    Labor Force Participation Rate (%) = (Labor Force / Working-Age Population) x 100
    
                                    The Labor Force includes both employed individuals and those who are unemployed but 
                                    actively seeking work.
                                    
                                    The Working-Age Population refers to the total number of individuals who are of working 
                                    age and considered to be part of the labor force potential. The definition of working 
                                    age can vary depending on the country and context.

Estimated Employed --> It represents the count of people who have jobs or are engaged in some form of employment.

Estimated Unemployment Rate (%) --> It is calculated by dividing the number of unemployed individuals by the total labor force 

                                    Estimated Unemployment Rate (%) = (Number of Unemployed / Labor Force) x 100

```python
# month of unemployment

# These code converts the 'Date' column to a datetime type, extracts the month as an integer, and adds a new column 
# with the corresponding three-letter month abbreviation.



import datetime as dt  # This line imports the datetime module, which provides classes and functions for working with dates and times.
import calendar        #  This line imports the calendar module, which includes various functions related to calendars, such as retrieving month names.

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)                 # This line converts the 'Date' column in the DataFrame df to a datetime data type using the pd.to_datetime() function from the pandas library. The dayfirst=True parameter specifies that the date format is in the day-month-year format, where the day appears before the month.
df['month_int'] = df['Date'].dt.month                                  #This line extracts the month component from the 'Date' column and assigns it to a new column called 'month_int' in the DataFrame df. The .dt accessor allows access to various date components, and .month returns the month as an integer.
df['month'] = df['month_int'].apply(lambda x: calendar.month_abbr[x])  # This line creates a new column called 'month' in the DataFrame df by applying a lambda function to the 'month_int' column. The lambda function uses calendar.month_abbr[x] to retrieve the three-letter abbreviation of the month based on the integer value in 'month_int'. The resulting abbreviations are stored in the 'month' column
```

```python
# visualize labor participation rate & unemployment rate in each month

df_EEE = df.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean()
df_EEE = pd.DataFrame(df_EEE).reset_index()
month = df_EEE.month
unemployment_rate = df_EEE['Estimated Unemployment Rate']
labor_participation_rate = df_EEE['Estimated Labour Participation Rate']

fig = go.Figure()             # This module provides the core functionality for creating interactive plots and visualizations using Plotly. It contains classes like Figure and various trace types such as Scatter, Bar, and Line. It is typically imported as go

fig.add_trace(go.Bar(x = month, y = unemployment_rate, name = 'umemployment_Rate'))                     # The fig.add_trace() method is used to add one or more traces (data series) to a Figure object in Plotly. Traces represent the data that you want to plot, such as scatter points, lines, bars, or other types of visual elements.
fig.add_trace(go.Bar(x = month, y = labor_participation_rate, name = 'Labor Participation Rate'))       # The trace object can be created using one of the trace classes from the plotly.graph_objects module, such as Scatter, Bar, Line, etc. Each trace class has its own set of attributes and parameters that allow you to customize the appearance and behavior of the trace

fig.update_layout(title='Unemployment Rate and Labor Participation Rate', xaxis={'categoryorder':'array', 'categoryarray':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']})
fig.show()
```
